```python
import numpy as np
import cv2 as cv

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_image=np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blank_image.shape
```

```python
plt.imshow(blank_image)
```

```python
cv.rectangle(blank_image, pt1=(348,10),pt2=(500,150),color=(0,255,0), thickness=10)
plt.imshow(blank_image)
```

```python
cv.rectangle(blank_image, pt1=(200,200),pt2=(300,300),color=(0,0,225), thickness=10)
plt.imshow(blank_image)
```

```python
cv.circle(img=blank_image, center=(100,100), radius=50,color=(244,0,0), thickness=5)
plt.imshow(blank_image)
```

```python
cv.circle(img=blank_image, center=(400,400), radius=50,color=(244,0,0), thickness=-1)
plt.imshow(blank_image)
```

```python
cv.line(blank_image,pt1=(0,0),pt2=(512,512),color=(102,255,255),thickness=5)
plt.imshow(blank_image)
```

```python
font=cv.FONT_HERSHEY_SIMPLEX
cv.putText(blank_image,text='Sachin',org=(10,500), fontFace=font,fontScale=4,color=(255,255,255),
          thickness=2, lineType=cv.LINE_AA)
plt.imshow(blank_image)
```

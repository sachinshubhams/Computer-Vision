```python
# import required packages
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab import files
```

```python
def generateContourThreshold(image,threshold_type):
  # create a binary thresholded image (use threshold=127)
  _, binary = threshold_type
  # show it
  plt.figure(figsize=(30, 30))
  plt.subplot(3, 2, 2)
  plt.title("Threshold")
  plt.imshow(binary, cmap="gray")
  # find the contours from the thresholded image (provide the output for every thresholding technique)
  contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
  # draw all contours
  image = cv2.drawContours(image, contours, -1, (0, 255, 0), 2)
  plt.subplot(3, 2, 3)
  plt.title("Contours")
  # show the image with the drawn contours
  plt.imshow(image)
  plt.show()
```

```python
uploaded = files.upload()
for fn in uploaded.keys():
    name=fn
image = plt.imread(name)
# convert to RGB
img_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
# convert to grayscale
img_grey = cv2.cvtColor(img_rgb,cv2.COLOR_RGB2GRAY)
```

```python
# read the image
image = plt.imread(name)

plt.figure(figsize=(30, 30))
plt.subplot(3, 2, 1)
plt.title("Original")
plt.imshow(image)
```

```python
generateContourThreshold(image,cv2.threshold(img_grey,200,255,cv2.THRESH_BINARY_INV))
```

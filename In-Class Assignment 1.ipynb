(d) Calculate the unnormalized histograms of the Red, Green, and Blue channels of the image and plot them using Matplotlib

```python
hist=cv.calcHist([img_rgb],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.xlim([0,256])
plt.ylim([0,18000])
plt.title('Channel R Histogram')
plt.xlabel('Bins')
plt.ylabel('# of Pixels')
plt.plot(hist)
```

(h) Use a binary threshold on your grayscale image and plot the result in the Jupyter Notebook using Matplotlib

```python
img_gray_th=cv.imread('cat_on_the_sofa.jpg',0)
```

```python
ret,thresh=cv.threshold(img_gray_th,125,255,cv.THRESH_BINARY)
```

```python
plt.imshow(thresh,cmap='gray')
plt.title('Binary Threshold Result ')
```

(i) Calculate the unnormalized histogram of the thresholded image and plot it using Matplotlib

```python
hist=cv.calcHist([thresh],channels=None,mask=None,histSize=[256],ranges=[0,256])
plt.xlim([0,256])
plt.title('Binary Threshold Image Histogram')
plt.xlabel('Bins')
plt.ylabel('# of Pixels')
plt.plot(hist)
```
